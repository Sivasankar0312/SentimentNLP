```python
import pandas as pd
import numpy as np
from sklearn.preprocessing import LabelEncoder
from tensorflow.keras.preprocessing.sequence import pad_sequences
from tensorflow.keras.preprocessing.text import Tokenizer
```

```python
import matplotlib.pyplot as plt
```

```python
import tensorflow as tf
```

```python
df_train=pd.read_csv("/content/train.txt",delimiter=";",names=['commants','sentiments'])
```

```python
df_train.head()
```

```python
df_valid=pd.read_csv("/content/val.txt",delimiter=";",names=['comments','sentiments'])
```

```python
df_valid
```

```python
df_test=pd.read_csv("/content/test.txt",delimiter=";",names=['comments','sentiments'])
```

```python
df_test
```

```python
df_train.sentiments.unique()
```

```python
df_train['sentiments'].value_counts()
```

```python
def re(text):
  text=text.replace("sadness","0")
  text=text.replace("anger","0")
  text=text.replace("fear","0")
  text=text.replace("joy","1")
  text=text.replace("love","1")
  text=text.replace("surprise","1")
  return text
```

```python
df_train['sentiments']=df_train['sentiments'].apply(lambda x:re(x))
```

```python
df_train['sentiments'].value_counts()
```

```python
df_test['sentiments']=df_test.sentiments.apply(lambda x:re(x))
```

```python
df_valid['sentiments']=df_valid.sentiments.apply(lambda x:re(x))
```

```python
df_valid['sentiments'].value_counts()
```

```python
vocab_size=10000
embeddingdim=64
oov_token='<oov>'
truncating_type='post'
padding_type='post'
max_length=128
```

```python
tokenizer=Tokenizer(num_words=vocab_size,oov_token=oov_token)
```

```python
tokenizer.fit_on_texts(df_train['commants'])
```

```python
word_index=tokenizer.word_index
```

```python
word_index
```

```python
train_sequences=tokenizer.texts_to_sequences(df_train['commants'])
```

```python
train_padsequences=pad_sequences(train_sequences,maxlen=max_length,padding=padding_type,truncating=truncating_type)
```

```python
df_train['sentiments']=pd.to_numeric(df_train['sentiments'])
```

```python
y_train=np.array(df_train['sentiments'])
```

```python
val_sequence=tokenizer.texts_to_sequences(df_valid['sentiments'])
val_padsequence=pad_sequences(val_sequence,maxlen=max_length,padding=padding_type,truncating=truncating_type)
```

```python
y_val=pd.to_numeric(df_valid['sentiments'])
```

```python
y_val
```

```python
model = tf.keras.Sequential([
    tf.keras.layers.Embedding(vocab_size, embeddingdim, input_length=max_length),
    tf.keras.layers.Bidirectional(tf.keras.layers.LSTM(128, return_sequences=True)),
    tf.keras.layers.Bidirectional(tf.keras.layers.LSTM(64)),
    tf.keras.layers.Dense(32, activation='relu'),
    tf.keras.layers.Dense(1, activation='sigmoid')
])
model.compile(loss='binary_crossentropy',optimizer='adam',metrics=['accuracy'])
```

```python
model.summary()
```

```python
num_epochs=10
history=model.fit(train_padsequences,y_train,epochs=num_epochs,validation_data=(val_padsequence,y_val))
```

```python
history.save('model')
```
